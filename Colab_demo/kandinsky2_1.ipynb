```python
from kandinsky2 import get_kandinsky2
```

```python
_prompt='''
**blue color palette** , **over watch mercy original character**, ~~overwatch mercy gained power and now full stack fairy,~~ 3d style, reality, realistic, far shot, outstanding graphics, ray traced, 8K, rtx on, trending on artstation, award winning, vivid colors, amazing contrast, focus on character, colorful, war theme, spacecraft backgrund, natural colors, sfw, symmetric body, badass, fire and ice on wings,**original**
'''
```

```python
model = get_kandinsky2('cuda', task_type='text2img', cache_dir='/tmp/kandinsky2', model_version='2.1', use_flash_attention=False)
```

```python
images = model.generate_text2img(
    _prompt,
    num_steps=100,
    batch_size=1,
    guidance_scale=4,
    h=768, w=768,
    sampler='p_sampler',
    prior_cf_scale=4,
    prior_steps="5"
)
```
